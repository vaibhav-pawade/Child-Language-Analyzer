```python
class analyser:
    length=''
    size_vocabalory=''
    rep_words=''
    ret_words=''
    errors=''
    pause=''
    def __init__(self):
        self.length=0
        self.size_vocabalory=0
        self.rep_words=0
        self.ret_words=0
        self.errors=0
        self.pause=0
        
    
    def __str__(self):
        #print("Length of the transcript : %s lines,Size of the vocabulary : %s words, Number of repetition for certain words or phrases : %s ,Number of retracing for certain words or phrases : %s ,Number of grammatical errors detected : %s, Number of pauses made : %s"%(self.length,self.size.vocabalory,self.rep_words,self.ret_words,self.errors,self.pause))
        print(self.length,self.size_vocabalory,self.rep_words,self.ret_words, self.pause,self.errors)
        
        
    def analyse_script(self, cleaned_file):
        folder=re.sub("-OP[0-9]+","",cleaned_file)
        f=open("Output\\"+folder+"-Output\\"+str(cleaned_file)+".txt",'r')
        self.length=len(f.readlines())
        f.close()
        f=open("Output\\"+folder+"-Output\\"+str(cleaned_file)+".txt",'r')
        st=f.read()
        f.close()
        w=re.findall("[^\s:\[\]\./\(\)!?][a-zA-Z]*",st)
        self.size_vocabalory=len(set(w))
        self.rep_words=len(re.findall("\[/\]",st))
        self.ret_words=len(re.findall("\[//\]",st))
        self.pause=len(re.findall("\(.\)",st))
        self.errors=len(re.findall("\[\*\]",st))
        #print(self.length,self.size_vocabalory,self.rep_words,self.ret_words, self.pause,self.errors)
        data=[cleaned_file,self.length,self.size_vocabalory,self.rep_words,self.ret_words, self.pause,self.errors]
        #analyse=pd.DataFrame([data],columns=['File_name','length','size_vocabalory','rep_words','ret_words','pause','errors'])
        return(data)
```

```python
analyseSLI=pd.DataFrame()
for i in range(1,11):
    sli=analyser()
    data_r=sli.analyse_script("SLI-OP"+str(i))
    d1=pd.DataFrame([data_r],columns=['File_name','length','size_vocabalory','rep_words','ret_words','pause','errors'])
    analyseSLI=d1.append(analyseSLI,ignore_index=True)
```

```python
analyseTD=pd.DataFrame()
for i in range(1,11):
    td=analyser()
    data_r=td.analyse_script("TD-OP"+str(i))
    d2=pd.DataFrame([data_r],columns=['File_name','length','size_vocabalory','rep_words','ret_words','pause','errors'])
    analyseTD=d2.append(analyseTD,ignore_index=True)
```
