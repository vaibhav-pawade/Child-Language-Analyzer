```python
a=re.findall("[*]CHI:\t.*\n\t.*|[*]CHI:\t.*\n",st)
a
```

```python
z=[]
for line in a:
    b=re.sub("\[[^//|/].*\]","",line)
    c=re.sub("[<|>]","",b)
    d=re.sub("&[\w*]|\+[\w*]","",c)
    e=re.sub("\n","",d)
    f=re.sub("\t","",e)
    if "(.)" not in d:
        g=re.sub("\(|\)","",f)
        z.append(g)
        print(g)
    else:
        z.append(f)
        print(f)
        
```
